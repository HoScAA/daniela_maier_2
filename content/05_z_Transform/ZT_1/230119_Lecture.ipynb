### Implentation of discrete time LTI-systems
---

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import TransferFunction, ZerosPolesGain
```

```python
sys1 = ZerosPolesGain([0,0],[.5j,-.5j],1,dt=True)
w, H = sys1.freqresp()
plt.plot(w,np.abs(H))
plt.grid()
plt.xlabel('Omega')
plt.ylabel('|H(e^(i*Omega)|')
```

```python
t,h = sys1.impulse()
plt.plot(t[:10],np.squeeze(h)[:10],'o')
plt.grid()
plt.xlabel('n')
plt.ylabel('h[n]')
```

```python
sys2 = TransferFunction([.5,.5],[1,0],dt=True)
w, H = sys2.freqresp()
plt.plot(w,np.abs(H))
plt.grid()
plt.xlabel('Omega')
plt.ylabel('|H(e^(i*Omega)|')
```

```python
t,h = sys2.impulse()
plt.plot(t[:5],np.squeeze(h)[:5],'o')
plt.grid()
plt.xlabel('n')
plt.ylabel('h[n]')
```
