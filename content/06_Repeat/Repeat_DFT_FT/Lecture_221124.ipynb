```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sp
from numpy.fft import fft, fft2, fftshift, ifftshift, ifft2, ifft
```

```python
f = np.array([1,0,1,1,0,0,0,1])
```

```python
c = fftshift(fft(f))
c.reshape(8,1)
```

```python
T = 2
f_s = 4
f_n = np.arange(-f_s/2,f_s/2,1/T)
plt.plot(f_n,np.abs(c),'o')
c_max = np.max(np.abs(c))
c[np.abs(c)<0.5*c_max]=0
plt.plot(f_n,np.abs(c),'o')
```
