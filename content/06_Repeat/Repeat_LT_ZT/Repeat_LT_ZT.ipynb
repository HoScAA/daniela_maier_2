Consider a discrete LTI-system with two poles $p_{1/2} = \pm 1$ and one zero $z_1 = 0$ (the gain is $2$).

- Sketch the pole-zero-map.
- Write down the transferfunction $H(z)$.
- Write down the corresponding difference equation.
- Perfrom a partial fraction decomposition in the transferfunction $H(z)$ and the fact, the $z$-transform of $f[n] = a^n$ is 
$$
F(z) = \frac{1}{1-\frac{a}{z}}
$$
to obtain analytically the impulse response $h[n]$. 

```python

sys = ZerosPolesGain([0],[1,-1],2,dt=True)
w, H = sys.freqresp()
plt.plot(w,np.abs(H))
plt.axis([0,np.pi,0,5])
```
