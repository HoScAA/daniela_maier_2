```python
import numpy as np
import matplotlib.pyplot as plt
```

Example: $R-C$-Circuit

In the time domain the equation reads
$$
RC\dot{U}_o + U_o = U_i
$$
and hence we obtain the transfer-function as
$$
H(\omega) = \frac{1}{RC(i\omega)+1}
$$

```python
R = 10
C = 0.001

w = np.linspace(0,10**3,10000)

H = 1/(R*C*1j*w+1)
plt.plot(w,np.abs(H))
plt.xscale('log')
plt.show()
plt.plot(w,np.angle(H))
```

```python
w = 800
H = 1/(R*C*1j*w+1)
H
t = np.linspace(0,2*np.pi/w*5,10000)
plt.plot(t,np.exp(1j*w*t).imag)
plt.plot(t,(H*np.exp(1j*w*t)).imag)
```
