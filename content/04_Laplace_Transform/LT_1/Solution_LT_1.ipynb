### Exercise 1
---

Calculate the Laplace Transform of the function
$$
f(t) = H(t)\cos(\alpha t)
$$
Your result sould read
$$
F(s) = \frac{s}{s^2+\alpha^2}
$$

### Solution 1

With
$$
\cos(\alpha t) = \frac{1}{2}\left(e^{i\alpha t} + e^{-i\alpha t}\right)
$$
we obtain
$$
F(s) = \frac{1}{2}\int_0^{\infty}\left(e^{-(s-i\alpha)t} + e^{-(s+i\alpha)t}\right)\mathrm{d}t = 
\frac{1}{2}\left(\frac{1}{s-i\alpha} + \frac{1}{s+i\alpha}\right) = \frac{s}{s^2+\alpha^2}
$$

### Exercise 2
---
Consider the ODE
$$
y''(t) -5y'(t) + 6y(t) = 0\qquad y(0) = 0,\, y'(0) = 2\,.
$$

(a) Determine $Y(s)$.

(b) Use partial fraction decomposition to determine $y(t)$ using a Laplace table.

(c) Plot $y(t)$.

### Solution 2
---
(a) We obtain
$$
(s^2Y(s)-2) - 5sY(s) + 6Y(s) = 0 \quad \Rightarrow \quad Y(s) = \frac{2}{s^2-5s+6} = \frac{2}{(s-2)(s-3)}
$$
(b) PFD yields
$$
Y(s) = \frac{2}{s-3}- \frac{2}{s-2}
$$
and hence
$$
y(t) = 2\left(e^{3t}-e^{2t}\right)
$$

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0,0.5,100)
y = 2*(np.exp(3*t)-np.exp(2*t))
plt.plot(t,y)
plt.grid()
plt.ylabel('y(t)')
plt.xlabel('t')
```

### Exercise 3
---
Consider the ODE
$$
y''(t) +4y'(t) + 5y(t) = e^{t}\qquad y(0) = 1,\, y'(0) = 2\,.
$$

(a) Determine $Y(s)$.

(b) Use partial fraction decomposition to determine $y(t)$ using a Laplace table. Use:
\begin{align}
\frac{1}{s^2+2\alpha s+\beta^2} & \quad\Laplace \quad \frac{e^{-\alpha t}}{\sqrt{\beta^2-\alpha^2}}\sin\left(\sqrt{\beta^2-\alpha^2}t\right) \\
\frac{s}{s^2+2\alpha s+\beta^2} & \quad\Laplace \quad \left[\cos\left(\sqrt{\beta^2 - \alpha^2}t\right) - \frac{\alpha}{\sqrt{\beta^2 - \alpha^2}} \sin\left(\sqrt{\beta^2 - \alpha^2} t\right)\right]e^{-\alpha t}
\end{align}

(c) Plot $y(t)$.

### Solution 3
---
(a) We obtain
$$
(s^2{Y(s)} - s-2) + 4 (sY(s) - 1) + 5Y(s) = \frac{1}{s-1}
$$
and hence
$$
Y(s)(s^2 + 4s + 5) = \frac{1}{s-1} + (s+6) \quad\Rightarrow \quad
Y(s) = \frac{1}{(s-1)(s^2+4s+5)} + \frac{(s+6)}{(s^2+4s+5)}\,.
$$
Finally we obtain:
$$
Y(s) = \frac{s^2+5s-5}{(s-1)(s^2+4s+5)}
$$
(b) PFD yields
$$
\frac{s^2+5s-5}{(s-1)(s^2+4s+5)} = \frac{A}{(s-1)} + \frac{Bs+C}{s^2+4s+5}
$$
and hence
$$
s^2 + 5s - 5 = A(s^2+4s+5) + Bs(s-1) +C(s-1) = s^2(A+B) +s(4A-B+C) +(5A-C)\,.
$$
We find that 
$$
A = \frac{1}{10}, \quad B = \frac{9}{10} \quad\mathrm{and}7\quad  C = \frac{55}{10}
$$
With the hint we obtain in the time domain
$$
y(t) = H(t)\left(\frac{1}{10}e^t + \frac{9}{10}\left(\cos(t)-2\sin(t)\right)e^{-2t} + \frac{55}{10}e^{-2t} \sin(t)\right)
$$
