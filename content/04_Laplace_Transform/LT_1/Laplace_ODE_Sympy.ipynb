### Using sympy to determine Laplace and Inverse Laplace transform

```python
from sympy import * 
t = symbols('t')
s = symbols('s')
```

#### Partial fraction decomposition

```python
apart(2/(s**2 - 5*s+6))
```

#### Solving ODEs

```python
f = Function('f')
dsolve(diff(f(t),t,2) -5*diff(f(t),t) + 6*f(t), f(t),ics={diff(f(t),t).subs(t,0) : 2 , f(0) : 0})
```

#### Laplace Transform

```python
init_printing = True
laplace_transform(Heaviside(t)*cos(3*t),t,s,noconds=True)
```

#### Inverse Laplace Transform

```python
inverse_laplace_transform(1/(s-1),s,t)
```
