# Lecture Laplace Transform 1

### Introduction
This section will introduce the **Laplace Transform**. After a short motivation we list the most important properties (and differences to the Fourier transform) of the Laplace transform. It will be shown, how we can transfer almost everything from the Fourier sections (LTI systems, impulse response function, transfer function, ...) to the Laplace theory. Finally we will show, how to solve LTI systems (described by differential equations) via Laplace transform.

#### Motivation
The Fourier transform $F(\omega)$ considered so far only exists, if the following condition holds
$$
\int_{-\infty}^{\infty} |f(t)|\, \mathrm{d}t < \infty\,.
$$
The function $f(t)$ hence has to vanish for large values of $t$ to make the integral converge. For some signals/functions the Fourier integral does not exist (i.e. the signal $f(t)=\Theta(t)$). We now modify the function in
the following way
$$
\tilde{f}(t) = e^{-\delta t}f(t),\qquad \delta \geq 0 \,.
$$
For $\delta\rightarrow 0$ we reobtain the original signal $f(t)$. If $\delta$ is large enough the integral
$$
\int_{-\infty}^{\infty} |\tilde{f}(t)|\, \mathrm{d}t = \int_{-\infty}^{\infty} |f(t)|e^{-\delta t}\, \mathrm{d}t
$$
will converge. The Fourier transform $\tilde{F}(\omega)$ is given by
$$
\tilde{F}(\omega) = \int_{-\infty}^{\infty}\tilde{f}(t) e^{-i\omega t} =  \int_{-\infty}^{\infty}f(t) e^{-(\delta+i\omega) t} = F(\delta + i\omega) 
$$
The Fourier transform $\tilde{F}(\omega)$ of the modified function $\tilde{f}$ is obtained from the Fourier transform of the function $f(t)$ via the replacement
$$
i\omega \rightarrow \delta + i\omega = s\,.
$$
The integral
$$
F(s) =  \int_{-\infty}^{\infty}f(t) e^{-s t} \mathrm{d}t
$$
is called the two-sided Laplace transform. As (almost all) signals start at $t=0$ the integration effectively runs from $0$ to $\infty$ only. The integral
$$
F(s) = \int_0^{\infty}f(t) e^{-s t} \mathrm{d}t
$$
is called the **Laplace transform**.

#### Properties of the Laplace transform

The Laplace transform $F(s)$ is defined as
$$
F(s) = \int_{0}^{\infty}f(t) e^{-st}\mathrm{d}t\,.
$$
This may also be written as
$$
\mathcal{L}\left(f(x)\right) = F(s)
$$
or 
$$
F(s)\,\,\Laplace\,\, f(t)
$$
The two main differences between the Fourier and the Laplace transform are given by
- The integration of the Laplace transform runs from $0$ to $\infty$ (instead of $-\infty$ to $\infty$.)
-  Instead of $i\omega$ we now have $s$. In some sense you may perform your calculation as $s$ was a real varible (but keep in mind that $s=i\omega + \delta$).

The inverse Laplace transform ist given by
$$
f(t) = \frac{1}{2\pi i}\int_{\delta+i\infty}^{\delta -i\infty}F(s) e^{st}\mathrm{d}s
$$
Note: In this lecture you won't perform an inverse Laplace transform by hand, as we will always look it up in a table.

**Example 1**: We calculate the Laplace transform of the function $f(t)=1$. We obtain
$$
F(s) = \int_{0}^{\infty}1 e^{-st}\mathrm{d}t = -\frac{1}{s}\left[e^{-st}\right]_0^{\infty} = \frac{1}{s}\,.
$$
**Example 2**: We calculate the Laplace transform of the function $f(t)=e^{-a t}$. We obtain 
$$
F(s) = \int_{0}^{\infty} e^{-a t} e^{-st}\mathrm{d}t = \int_{0}^{\infty} e^{-(a+s) t} \mathrm{d}t  =  -\frac{1}{s+a}\left[e^{-st}\right]_0^{\infty} = \frac{1}{s+a}
$$

|   $e^{iat} $                | $   \frac{s+ia}{s^2+a^2} $  |
|   $\cos(at) $               | $ \frac{s}{s^2+a^2} $       |
|   $ \sin(at) $              | $\frac{a}{s^2+a^2} $        |     
|   $ 1$              | $ \frac{1}{s}$              |
|   $ e^{at}   $ | $ \frac{1}{s- a} $        |
|   $ \frac{t^{n-1}}{(n-1)!}e^{ at}  $ | $ \frac{1}{(s- a)^n}$     |


Important properties of the Laplace transform are listed in the table below:

|     Property            | Function $f(t)$             | Fourier transform $F(S)$   |
|------------------|-----------------------------|---------------------------------|
| Linearity        | $a_1f_1(t) + a_2f_2(t) $    | $ a_1F_1(S) + a_2F_2(s) $ |

| Differentation   | $f^{(n)}(t)$               | $  s^nF(s) - \sum\limits_{k=0}^{n-1}s^k f^{(n-1-k)}(0)$ |


#### Solution of ordinary differential equation via Laplace transform
We consider the problem of an LTI system, given by the differential equation
$$
Ly(t) = x(t)
$$
with appropriate initial conditions (depending on the order of the 
differential operator $L$). We can now perform the same procedure as in the Fourier section to solve this differential equation using the Laplace
transform. There is one big difference/advantage: Using Laplace transform we obtain in one shot the solution to the initial value problem (i.e. the solution contains the homogeneous and the particular solution with fixed coefficients).

As in the Fourier section the second step is to find Y(s) and the third step is to transform back to the time domain. This will (in almost all cases) lead to the problem of a partial fraction decomposistion. The following table contains the important rules for partial fraction decompositions (here we list the rules for real zeros only). It is supposed to have a fraction of the form
$$
\frac{g(x)}{h(x)}, \quad \mathrm{with}\quad h(x) = (x-x_1)^{r_1}\cdots
(x-x_n)^{r_n}
$$

| Zero type of h(x)           | Ansatz for PFD            |
|-----------------------------|-----------------------------|
|   zero of multiplicity 1, i.e. $(x-x_i)$     | $ \frac{a_{i1}}{x-x_i} $ |              
|   zero of multiplicity r, i.e.  $(x-x_i)^{r_i}$  | $ \frac{a_{i1}}{x-x_i}+ ... + \frac{a_{ir_i}}{(x-x)^{r_i}}$ |              


**Example 1** : We consider the initial value problem
$$
y'(t) + y(t) = \theta(t)e^{-t}, \quad y(0)=0
$$
We transform and obtain
$$
sY(s) + Y(s) = \frac{1}{s+1} \quad \Rightarrow \quad Y(s) = \frac{1}{(s+1)^2}
$$
Via the Laplace transform table from above we obtain the inverse Laplace transform as
$$
y(t) = \theta(t) te^{-t}
$$

**Example 2** : We consider the initial value problem
$$
y''(t)+ 5y'(t)+ 6y(t) = 0, \quad y(0)=2, y'(0) = 3
$$
With the initial conditions we obtain
$$
(s^2Y(s)-2s-3) + 5(sY(s)-2) +6Y(s) = 0 \quad \Rightarrow \quad 
Y(s)(s^2+5s+6) = 2s+13
$$
This leads to
$$
Y(s) = \frac{2s+13}{(s+3)(s+2)}
$$
Partial fraction decomposition 
$$
 \frac{2s+13}{(s+3)(s+2)} = \frac{A}{s+2}+\frac{B}{s+3} \quad \Rightarrow \quad
 2s+13 = A(s+3) + B(s+2) = s(A+B) + (3A+2B)
$$
and hence $A=9$ and $B=-7$. So we obtain
$$
y(t) = 9e^{-2t}-7e^{-3t}
$$
